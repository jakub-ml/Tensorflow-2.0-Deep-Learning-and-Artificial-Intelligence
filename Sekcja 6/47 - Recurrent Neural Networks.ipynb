# **GRU** and **LSTM**
### W zwykłych RNN występuje też problem vanishing gradient descent, dlatego RNN jest słabe jak chodzi o zapamiętywanie dłuższych sekwencji czasowych. Dlatego jeśli zapytaliśmy RNN w konktekście NLP co znajdowało się na początku artykułu który kazaliśmy mu przeczytać, nie wiedziałby bo zdąrzyłby zapomnieć. W tym przypadku zastosowanie Relu jako funckji aktywacji nie wystarczy. Zamiast tego korzystamy z innej architektury.


* GRU - uproszczona wersja LSTM - pozwala zapamiętywać dłuższe sekwencje dzięki mechanizmowi zapamiętywania i zapominania komkretnych cech, mechanizm tego czy chcemy coś pamiętać czy nie jest zwykłą binarną klasyfikacją.

* LSTM - lepsze od GRU, ma więcej parametrów, bardziej złożone

Więc jeśli chcemy żeby model przewidywał coś na podstawie 5 próbek wstecz to RNN może wystarczyć, ale jeśli będziemy chcieli wykorzystać ich 20, RNN może sobie nie dac rady i będzie trzeba wykorzystać LSTM. Jeśli liczba przektory jakaś wartośc np. 30, obie metody mogą nie dać sobie rady.
