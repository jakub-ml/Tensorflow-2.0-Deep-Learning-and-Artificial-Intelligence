### Ten notatnik polega na tym że uczymy model przewidywania wyników funkcji cos(2x_1)+cos(3x_2)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
zakres = np.linspace(-3,3,100)
x_s = np.array(np.meshgrid(zakres, zakres)).T.reshape(-1, 2)
x, y = x_s[:,0], x_s[:,1]
z=np.cos(3*x)+np.cos(2*y)
```

```python
 # Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(x, y, z, color = "green")
plt.title("simple 3D scatter plot")

# show plot
plt.show()
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(2, )),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(1)
])
```

```python
opt=tf.keras.optimizers.Adam(0.01)
model.compile(optimizer=opt,
              loss='mse')
```

```python
r = model.fit(x_s, z, epochs=100)
```

```python
plt.plot( r.history['loss'])
plt.show()
```

```python
zakres = np.linspace(-5,5,50)
x_s = np.array(np.meshgrid(zakres, zakres)).T.reshape(-1, 2)
x, y = x_s[:,0], x_s[:,1]
z=model.predict(x_s)
```

```python
 # Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(x, y, z, color = "green")
plt.title("simple 3D scatter plot")

# show plot
plt.show()
```
