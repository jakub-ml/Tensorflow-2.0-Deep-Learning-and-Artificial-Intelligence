### Architektura neurona jest oparta na prawdziwych neuronach. Do neurona są skierowane połączenia z sygnałami które są mnożone przez wagę znajdującą się na połączeniu, im większa jest wartość tym większe znaczenie sygnału dla neurona, sume wych wartości przekazujemy do funkjci aktywacji, która sumę wyznałów zamienia na informacjie jak bardzo sam neuron jest aktywny.

### Wykres pokazujący co wykres loss function może nam powiedzieć o learning rate, warto zwrócić wagę na **learning rate scheduling** - czyli zmiana wartości learning rate w którejś epoce uczenia modelu.
